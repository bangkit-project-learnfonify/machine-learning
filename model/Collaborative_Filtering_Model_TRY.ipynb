```python
import numpy as np
import pandas as pd
import warnings
import tensorflow as tf
from keras.models import load_model
```

```python
# # use GDrive
# from google.colab import drive
# drive.mount('/content/drive')
# path_to_model = '/content/drive/My Drive/model/recommendation_rating_model.h5'

# use local directory
path_to_model = '/content/recommendation_rating_model.h5'
```

```python
# load model
model = load_model(path_to_model)

# summarize model
model.summary()
```

```python
#Making recommendations
course_id = range(1,322) # input course IDs
course_arr = np.array(course_id) #get all course IDs
```

```python
user_id = 190 # input user id
user = np.array([user_id for i in range(len(course_id))]) # check for user
pred = model.predict([course_arr, user])
pred
```

```python
pred = pred.reshape(-1) #reshape to single dimension
pred_ids = (-pred).argsort()[0:10] # predict 10 courses
```
